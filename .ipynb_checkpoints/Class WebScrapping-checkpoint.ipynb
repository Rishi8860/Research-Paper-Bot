```python
import requests as rq
import bs4

class Books_toscrape_com():
    def __init__(self,no_of_pages):
        self.all_pages=[f"https://books.toscrape.com/catalogue/page-{i}.html" for i in range(1,no_of_pages+1)]
        self.no_of_pages=no_of_pages
    def __page_book(self,website):
        data=rq.get(website)
        print(data.status_code)
        if (data.status_code==200):
            data=bs4.BeautifulSoup(data.text,'html.parser')
            books=data.find_all(class_='product_pod')
            title=[]
            for i in books:
                title.append(i.h3.text)
            print(len(title))
            return title
        else:
            return []
    def Get_All_Books(self):
        books=[]
        for i in self.all_pages:
            books.extend(self.__page_book(i))
        return books
    def Get_Books_Of_Page(self,n):
        page=f"https://books.toscrape.com/catalogue/page-{n}.html"
        return self.__page_book(page)
    def Get_Categories(self):
        data=rq.get('https://books.toscrape.com/')
        if data.status_code==200:
            data=bs4.BeautifulSoup(data.text,'html.parser')
            side_bar=data.find_all(class_='sidebar')
            categories=[]
            for i in side_bar:
                j=i.find_all('li')
                for k in j:
                    categories.append((k.a.text.strip(),'https://books.toscrape.com/'+k.a['href']))
            return categories[1:]
        else:
            return []
    '''def __del__(self):
        print('Destructor called, Memory Released.')
    '''
```
